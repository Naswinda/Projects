```python
import pandas as pd
import plotly.express as px
import plotly.graph_objects as go
```

# **1. Get The Data**

```python
df = pd.read_csv("https://raw.githubusercontent.com/Naswinda/nida/main/dads5001/Plotly/disney_plus_titles.csv")
df.head()
```

```python
df.info()
```

# **2. Clean and Analyze DATA**

```python
data_exploded = df.assign(listed_in=df['listed_in'].str.split(', ')).explode('listed_in')
data_exploded
```

```python
data_counts = data_exploded['listed_in'].value_counts().reset_index()
data_counts.columns = ['Category', 'Count']
data_counts
```

# **3. Get The Pie Chart**

```python
# Filter for groups with count less than 200
filtered_data = data_counts[data_counts['Count'] >= 200]

# Calculate total count for remaining groups
other_count = data_counts[data_counts['Count'] < 200].sum()['Count']

# Create new DataFrame
new_df = pd.DataFrame({
    'Category': pd.concat([filtered_data['Category'], pd.Series(['Other'])]),
    'Count': pd.concat([filtered_data['Count'], pd.Series([other_count])])
})

print(new_df)
```

```python
fig = go.Figure(data=[go.Pie(labels=new_df['Category'],
                              values=new_df['Count'],
                              pull=[0.1, 0, 0, 0])])
fig.update_layout(title='The GENRES on the Disney+ in 2021')
fig.show()
```
