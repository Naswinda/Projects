**3.นัดที่แมนยูแพ้ใครเป็นผู้ตัดสินบ้าง** (ปฏิเสธไม่ได้ว่าในการแข่งขันฟุตบอลนั้น ผู้ตัดสินนั้นเป็นตัวแปรสำคัญในการแข่งขัน จึงต้องการทราบชื่อผู้ตัดสินในเกสม์ที่แพ้ เพื่อวิเคราะห์ว่าอาจจะมีผู้ตัดสินคนไหนที่ให้การตัดสินไม่เป็นธรรม)

ผลการแข่งขันแต่ละนัดของ Man United

```python
manu_matches = pl[(pl["HomeTeam"] == "Man United") | (pl["AwayTeam"] == "Man United")]
man_u = manu_matches.loc[:, ["HomeTeam","AwayTeam","FTHG","FTAG","FTR"]]
man_u
```

ผู้ตัดสินในแต่ละนัดที่แมนยูแพ้

```python
referee = manu_matches.loc[:, ["HomeTeam","AwayTeam","FTHG","FTAG","FTR","Referee"]]
ref_away = referee[(referee["AwayTeam"] == "Man United") & (referee["FTR"] == "H")]
ref_home = referee[(referee["HomeTeam"] == "Man United") & (referee["FTR"] == "A")]
manu_loss = pd.concat([ref_away, ref_home])
sort_manu_loss= manu_loss.sort_values(by=['FTHG', 'FTAG'], ascending=[False, False])
sort_manu_loss
```

**4.ทีมเยือนที่ชนะด้วยสกอร์ห่างที่สุด 3 อันดับ**

```python
away_team = pl.loc[pl["FTR"] == "A", ['HomeTeam', 'AwayTeam', 'FTHG', 'FTAG']]
away_team['FTAG-FTHG'] = away_team['FTAG'] - away_team['FTHG']
max = away_team.nlargest(5, 'FTAG-FTHG')
max
```

**5.เปอร์เซนต์การแข่งขันชนะ ของเจ้าบ้าน และ ทีมเยือน**

```python
total_matches = len(pl)

home_team = pl.loc[pl["FTR"] == "H"]
home_win = len(home_team)
per_home_win = np.around((home_win / total_matches) * 100, 2)

away_team = pl.loc[pl["FTR"] == "A"]
away_win = len(away_team)
per_away_win = np.around((away_win / total_matches) * 100, 2)

print(f'HomeTeam wins {(per_home_win)}% \nAwayTeam wins {per_away_win}% ')
```
